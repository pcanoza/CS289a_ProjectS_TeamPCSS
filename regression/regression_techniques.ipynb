## Linear Regression
Now, we set up the code to do linear regression.

### Featurization
Here we generate a complex-valued matrix that captures the Fourier featurization. The code fixes some bugs and is a runtime improvement over the fourier featurization of the Team SKEP's provided code by leveraging numpy functions.

```python
def fourier_sequence(t, n_fourier):
    #seq = (np.arange(n_fourier) + 1)
    k = 2 * np.pi * np.arange(n_fourier)/ n_fourier
    #seq = np.linspace(0, n_fourier, n_fourier) 
    #return np.power(np.e, 2 * np.pi * 1j * (t * seq ))
    #return np.power(np.e, 2 * np.pi * 1j * (t / seq )) / np.sqrt(2*n_fourier + 1) 
    return np.power(np.e, 1j * (t * k ))
    # return np.power(np.e, 2 * np.pi * 1j * (t * seq)) 
def fourier_features(n_fourier, t_range):
    mat = np.zeros((len(t_range), n_fourier), dtype=complex)
    for i in range(len(t_range)):
        mat[i] = fourier_sequence(t_range[i], n_fourier) 
    return mat.T
```

### Least Square Methods

We will resort to simple Least Square methods that we have learned in class, namely normal OLS and Ridge. Notice that since we are dealing with complex valued matrices, we need to use Hermitian conjugates instead of transposes.

```python
# Normal, matrix-wise OLS
# Notice we need to use Hermitian conjugate instead of transpose
# since we are using complex valued matrices
# Returns learned weights and the residual
def ols(A, y):
    w = np.linalg.solve(A.H @ A, (A.H @ y).T)
    res = np.linalg.norm(y - A @ w)
    return w, res

# Matrix-wise Ridge regression
# If lmbda = 0, becomes OLS
# Returns learned weights and the residual
def ridge(A, y, lmbda = 10):
    w = np.linalg.solve(A.H @ A + lmbda * np.eye(A.shape[1]), (A.H @ y).T)
    res = np.linalg.norm(y - A @ w)
    return w, res
```

### Training/Testing Loop on  X, Y data
Here, we learn the coefficients for a given complex Fourier featurization of the planets. We train on the 5 day intervals of the noisy data, report the training error.  We then test on the 1 day intervals of the clean data. Note that if we are training on the moon, we are forced to use the clean data due to the sheer amount of noise, as mentioned before.
<br><br>
Notice that we cast the data into a complex number, so we have $$z = x_{planet} + j y_{planet} $$
And so we are essentially trying to learn the coeficients for the problem
$$ z(t) \approx \sum_i a_i e^{j k_i t} $$
which is a generic way to write quasi-periodic motion using the Fourier features.
<br> <br>
The user can change the subset of planets the loop runs over. They can also experiment with the lambda parameter for Ridge to see how/if regularization can help.

```python
##### Training set up #####
planets = BODY_NAMES             # Subset of planets to train...BODY_NAMES for all
lmbdas = [200, 0]                #Lambda parameter for ridge. lmbda = 0 is OLS, and should always be on the list
n_times = [1000, 2000]           # Must at least capture the length of the "quasi-cycle" so model is accurate
plot_fig = True                  # Set true if want to see visualizations
print_1D = False                  # Set to true if want to view 1D path...else, 5D path

coefs = []                       # Collecting the coefficients

##### Main Loop #####
for planet in planets:
    planet_cmplx = np.array(final_data[planet + "_x"] + 1j*final_data[planet + "_y"]).T
    planet_clean = np.array(clean_data[planet + "_x"] + 1j* clean_data[planet + "_y"])
    if planet=="moon":
        planet_cmplx = planet_clean
    planet_clean_1D = np.array(clean_data_1D[planet + "_x"] + 1j* clean_data_1D[planet + "_y"])

    #-- Time range loop ---
    for n_time in n_times:
        n_fourier = n_time     # Can change number of fourier features
        if plot_fig:
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
        print("===== Planet: %s, n_time: %s, n_fourier: %s =====" % (planet, n_time, n_fourier))
        
        #-- Lmbda range loop ---
        for lmbda in lmbdas:
        
            print("*Lambda = %s" % lmbda)
            # Least Squares Regression
            t_range = np.arange(n_time)
            mat_fourier = np.matrix(fourier_features(n_fourier, t_range)).T
            coef, train_loss = ridge(mat_fourier, planet_cmplx[:n_time], lmbda)
            print("---> Training loss: %s" % train_loss)
            
            # Calculate 1 Day test loss
            t_range_test = np.arange(n_time * 5)/5
            mat_fourier_test = np.matrix(fourier_features(n_fourier, t_range_test)).T
            test_path = mat_fourier_test @ coef
            test_loss = np.linalg.norm(test_path - planet_clean_1D[:n_time * 5])
            print("---> Test loss: %s" % test_loss)
        
            # Visualize Learned Weights
            if plot_fig:
                ax1.set_title("Fourier Weights: " + planet)
                ax1.plot(2 * np.pi * np.arange(n_fourier)/ n_fourier, np.abs(coef), label="lambda=%s"%lmbda)
        #-- End lmbda loop ---
            
        # Visualize Learned Path of last lmbda's coefficients
        if plot_fig:
            pred_path = mat_fourier @ coef
            ax2.set_title("Predicted vs Clean Path: " + planet)
            
            if print_1D:
                ax2.plot(planet_clean_1D[:n_time].real, planet_clean_1D[:n_time].imag, label = "clean_data")
                ax2.plot(test_path.real, test_path.imag, label = "1D prediction, lmbda=%s" %lmbda)
            else:
                ax2.plot(planet_clean[:n_time].real, planet_clean[:n_time].imag, label = "clean_data")
                ax2.plot(pred_path.real, pred_path.imag, label = "5D prediction, lmbda=%s" %lmbda)
            ax1.set_xlabel("Frequency")
            ax1.set_ylabel("abs(coef)")
            ax2.set_xlabel("X (AUs)")
            ax2.set_ylabel("Y (AUs))")
            ax1.legend()
            ax2.legend()
            plt.show()
        
    #-- End time loop ---
        
    # Save last learned coeficient
    coefs.append(coef) 
        
# Zip together learned coefficients in dictionary         
learned_coefs = dict(zip(planets, coefs))
```

### Peaks and Ptolemy
To see how the regression is related to Ptolemy's original conception of the model, we plot the path for the 2 largest coeficients of the fourier series that would correspond to the orbit and the epicycle, as well as tacking on the 0th fourier coeficient to represent the deferent. 
<br> <br>
Unfortunately, we do see very large deviations from ideal behaviour, as we suspect that there are many more than 2 coeficients that contribute to the final shape when we do a generic regression technique like this.  You can try to adjust the number of the largest coeficients to plot. 

```python
planet = "sun"   # Choose a planet
peak_n = 2          # Choose nth largest coeficients

# Retrieve coef and generate features
planet_clean = np.array(clean_data[planet + "_x"] + 1j* clean_data[planet + "_y"])
coef = learned_coefs[planet]
n_time = len(coef)
t_range = np.arange(n_time)
mat_fourier = np.matrix(fourier_features(n_time, t_range)).T

# Find peaks
a = np.ndarray.flatten(-np.abs(coef))
peaks = np.ndarray.flatten(a.argsort()[:,:peak_n])

# Add the nth largest features
pred_peak = 0
for i in range(peak_n):
    p = peaks[0,i]
    pred_peak += mat_fourier[:,p] * coef[p]
pred_peak += mat_fourier[:,0] * coef[0]

# Plot peak vs clean data
plt.plot(pred_peak.real, pred_peak.imag, label = "Peak Prediction") 
plt.plot(planet_clean[:n_time].real, planet_clean[:n_time].imag, label = "Clean Data")
plt.legend()
plt.show()
```

### Predicting Forward
To see how our model does, we can try to predict forward into the future and see how it does.  What becomes immediately evident after playing trying this out is that the Fourier features have a very large tendency to overfit, and for some reason do not capture the drift of the orbits as time passes by.

```python
planet = "venus"    # Choose a planet
n_forward =  100   # Choose how many time steps ahead to try and predict

planet_clean = np.array(clean_data[planet + "_x"] + 1j* clean_data[planet + "_y"])
coef = learned_coefs[planet]
n_fourier = len(coef)
n_time = n_fourier
t_range = np.arange(n_time + n_forward)
mat_forward = np.matrix(fourier_features(n_fourier, t_range[n_time:])).T
pred_cmplx = mat_forward @ coef
plt.title("Forward Prediction: %s steps forward" % n_forward)
plt.plot(pred_cmplx.real, pred_cmplx.imag, label = "prediction")
plt.plot(planet_clean[n_time:n_time+n_forward].real, 
         planet_clean[n_time:n_time+n_forward].imag, label = "clean_data")
plt.legend()
plt.show()

pred_loss = np.linalg.norm(pred_cmplx - planet_clean[n_time:n_time+n_forward])
print("Prediction loss: %s"%pred_loss )
```

### Learning the Z coordinates
We essentially copy and paste the same method as before, but now we only care about the real part of the Fourier featurization.  Note that Z coordinates are especially susceptible to noise, and thus really do not work great for this unrefined methodology. You can try out values below, and toggle between training on clean vs dirty data.

```python
##### Training set up #####
planets = BODY_NAMES            # Subset of planets to train...BODY_NAMES for all
lmbdas = [0]                     #Lambda parameter for ridge. lmbda = 0 is OLS, and should always be on the list
n_times = [100, 500]             # Must at least capture the length of the "quasi-cycle" so model is accurate
plot_fig = True                  # Set true if want to see visualizations
print_1D = False                 # Set to true if want to view 1D path...else, 5D path
train_clean = False              # We can see results if we train on clean data
coefs_z = []                     # Collecting the coefficients

##### Main Loop #####
for planet in planets:
    planet_z = np.array(final_data[planet + "_z"])
    planet_clean_z = np.array(clean_data[planet + "_z"])
    if train_clean:
        planet_z = planet_clean_z
    planet_clean_1D_z = np.array(clean_data_1D[planet + "_z"])

    #-- Time range loop ---
    for n_time in n_times:
        n_fourier = n_time     # Can change number of fourier features
        if plot_fig:
            fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
        print("===== Planet: %s, n_time: %s, n_fourier: %s =====" % (planet, n_time, n_fourier))
        
        #-- Lmbda range loop ---
        for lmbda in lmbdas:
        
            print("*Lambda = %s" % lmbda)
            # Least Squares Regression
            t_range = np.arange(n_time)
            mat_fourier = np.matrix(fourier_features(n_fourier, t_range)).T.real    #only using real
            coef_z, train_loss = ridge(mat_fourier, planet_z[:n_time], lmbda)
            print("---> Training loss: %s" % train_loss)
            
            # Calculate 1 Day test loss
            t_range_test = np.arange(n_time * 5)/5
            mat_fourier_test = np.matrix(fourier_features(n_fourier, t_range_test)).T.real  #only using real
            test_path = mat_fourier_test @ coef_z
            test_loss = np.linalg.norm(test_path - planet_clean_1D_z[:n_time * 5])
            print("---> Test loss: %s" % test_loss)
        
            # Visualize Learned Weights
            if plot_fig:
                ax1.set_title("Fourier Weights: " + planet)
                ax1.plot(2 * np.pi * np.arange(n_fourier)/ n_fourier, np.abs(coef_z), label="lambda=%s"%lmbda)
        #-- End lmbda loop ---
            
        # Visualize Learned Path of last lmbda's coefficients
        if plot_fig:
            pred_path = mat_fourier.real @ coef_z.real
            ax2.set_title("Predicted vs Clean Path: " + planet)
            
            if print_1D:
                ax2.plot(planet_clean_1D_z[:n_time], label = "clean_data")
                ax2.plot(test_path, label = "1D prediction, lmbda=%s" %lmbda)
            else:
                ax2.plot(planet_clean_z[:n_time], label = "clean_data")
                ax2.plot(pred_path, label = "5D prediction, lmbda=%s" %lmbda)
            ax1.legend()
            ax2.legend()
            plt.show()
        
    #-- End time loop ---
        
    # Save last learned coeficient
    coefs_z.append(coef_z) 
        
# Zip together learned coefficients in dictionary         
learned_coefs_z = dict(zip(planets, coefs_z))
```

# Regression with Moon Phases

Note that moon phase is just time series data of the moon phase angle.  Thus, we can use our usual 1D techniques. Let's see what the data looks like

```python
fig, ax = plt.subplots()
ax.set_title("Example of Moon Phase")
ax.plot(final_data["moon_phase"][:100])
ax.set_xlabel("time (5D)")
ax.set_ylabel("phase angle")
plt.show()
```

Simple 1D Fourier features should do! Let's borrow code from Homework 1...

```python
from sklearn.linear_model import LinearRegression
def solve_ls(phi, y):
    
    LR = LinearRegression(fit_intercept=False, normalize=False)

    LR.fit(phi, y)
    coeffs = LR.coef_
    
    loss = np.mean((y- phi@coeffs)**2)
    return coeffs, loss
def featurize_fourier(x, d, normalize = False):
    assert (d-1) % 2 == 0, "d must be odd"
    max_r = int((d-1)/2)
    n = len(x)
    A = np.zeros((n, d))
    A[:,0] = 1
    for d_ in range(1,max_r+1):
        A[:,2*(d_-1)+1] =  np.sin(d_*x*np.pi)
        A[:,2*(d_-1)+2] =  np.cos(d_*x*np.pi)
    
    if normalize:
        A[:,0] *= (1/np.sqrt(2))
        A *= np.sqrt(2)
    return A
```

Let's run regression! Notice that we need to scale down the time range for this featurization, since we are using integer sized frequencies.

```python
# Train
n_time = 100
n_range = np.arange(n_time)/ n_time
moon_train = np.array(final_data["moon_phase"][:n_time])
feat = featurize_fourier(n_range, n_time + 1)
coef, loss = solve_ls(feat, moon_train)
print("Training loss: %s" % loss)
```

```python
fig, ax = plt.subplots()
ax.set_title("Pred vs Actual Moon Phase")
ax.set_xlabel("time (5D)")
ax.set_ylabel("phase angle")
plt.plot((feat @ coef) [:100], label="predicted")
plt.plot(moon_train[:100], label = "actual")
plt.legend()
plt.show()
```
